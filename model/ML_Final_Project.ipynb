```python
import pandas as pd
```

```python
traffic_df = pd.read_csv('./traffic_data.csv')
traffic_df['Date_and_Time'] = pd.to_datetime(traffic_df['Date_and_Time'])

traffic_df
```

```python
weather_df = pd.read_csv('./open-meteo-6.22N106.78E0m(1).csv')
weather_df['time'] = pd.to_datetime(weather_df['time'])

weather_df
```

```python
traffic_df = pd.merge(traffic_df, weather_df, left_on='Date_and_Time', right_on='time')
traffic_df
```

```python
numeric_col = [
    'In_Motorcycle',
    'In_Car',
    'In_BusTruck',
    'In_Total',
    'In_PCE',
    'In_Velocity',
    'In_VCRatio',
    'Out_Motorcycle',
    'Out_Car',
    'Out_BusTruck',
    'Out_Total',
    'Out_PCE',
    'Out_Velocity',
    'Out_VCRatio',
    'Total_Motorcycle',
    'Total_Car',
    'Total_BusTruck',
    'Total_Total',
    'Total_PCE',
    'Total_Velocity',
    'Total_VCRatio',
    'temperature_2m (°C)',
    'relative_humidity_2m (%)',
    'precipitation (mm)',
    'wind_speed_10m (km/h)',
    'cloud_cover (%)',
    'surface_pressure (hPa)'
]

def object_to_numeric(df, col):
  df[col] = df[col].astype(str).str.replace(',', '')
  df[col] = pd.to_numeric(df[col], errors='coerce')

for col in numeric_col:
  object_to_numeric(traffic_df, col)

traffic_df = traffic_df.drop(columns=['time'])

traffic_df
```

```python
traffic_df['Date_and_Time'] = pd.to_datetime(traffic_df['Date_and_Time'])

traffic_df['Hour'] = traffic_df['Date_and_Time'].dt.hour
traffic_df['Day'] = traffic_df['Date_and_Time'].dt.day
traffic_df['Month'] = traffic_df['Date_and_Time'].dt.month

traffic_df
```

```python
from sklearn.impute import SimpleImputer
from sklearn.mixture import GaussianMixture

traffic_df_imputed = traffic_df.copy()

temp_imputer = SimpleImputer(strategy='median')
df_temp = pd.DataFrame(temp_imputer.fit_transform(traffic_df[numeric_col]),
                      columns=numeric_col)

gmm = GaussianMixture(n_components=2, random_state=42)
clusters = gmm.fit_predict(df_temp)

for col in numeric_col:
  for cluster in [0, 1]:
    cluster_mask = (clusters == cluster) & (~traffic_df[col].isna())
    cluster_values = traffic_df.loc[cluster_mask, col]

    cluster_median = cluster_values.median()

    missing_in_cluster = traffic_df[col].isna() & (clusters == cluster)
    traffic_df_imputed.loc[missing_in_cluster, col] = cluster_median

traffic_df_imputed.info()
```

```python
from sklearn.model_selection import train_test_split

X_features = [
    'relative_humidity_2m (%)',
    'precipitation (mm)',
    'cloud_cover (%)',
    'Hour',
    'Day',
    'Month'
]

X = traffic_df_imputed[X_features]
y = traffic_df_imputed['Total_Total']

X_train, X_test, y_train, y_test = train_test_split(X, y)

X_train, X_test, y_train, y_test
```

```python
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.svm import SVR
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, root_mean_squared_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.feature_selection import SelectKBest, f_regression
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

rf_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('feature_selection', SelectKBest(score_func=f_regression)),
    ('model', RandomForestRegressor(random_state=42))
])

rf_params = {
    'feature_selection__k': ['all'],
    'model__n_estimators': [50, 100, 200],
    'model__max_depth': [None, 10, 20],
    'model__min_samples_split': [2, 5, 10]
}
```

```python
xgb_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', XGBRegressor(random_state=42))
])

xgb_params = {
    'model__n_estimators': [50, 100, 200],
    'model__learning_rate': [0.01, 0.1, 0.2],
    'model__max_depth': [3, 6, 9],
    'model__subsample': [0.8, 1.0],
    'model__colsample_bytree': [0.8, 1.0]
}
```

```python
svm_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', SVR())
])

svm_params = {
    'model__kernel': ['poly'],
    'model__degree': [2, 3],
    'model__C': [0.1, 1, 10],
    'model__epsilon': [0.01, 0.1, 0.5]
}
```

```python
lr_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LinearRegression())
])

lr_params = {
    'model__fit_intercept': [True, False]
}
```

```python
pipelines = [
    ('Random Forest', rf_pipeline, rf_params),
    ('XGBoost', xgb_pipeline, xgb_params),
    ('SVM', svm_pipeline, svm_params),
    ('Linear Regression', lr_pipeline, lr_params)
]

def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    print(f"Y Predictions: {y_pred}")
    print(f"MAE: {mean_absolute_error(y_test, y_pred):.2f}")
    print(f"RMSE: {root_mean_squared_error(y_test, y_pred):.2f}")
    print(f"R²: {r2_score(y_test, y_pred):.2f}")
    return y_pred

best_models = {}
for name, pipeline, params in pipelines:
    print(f"\n=== Training {name} ===")

    grid_search = GridSearchCV(
        estimator=pipeline,
        param_grid=params,
        cv=5,
        scoring='neg_mean_squared_error',
        n_jobs=-1,
        verbose=1
    )

    grid_search.fit(X_train, y_train)

    best_model = grid_search.best_estimator_
    best_models[name] = best_model

    print(f"Best params: {grid_search.best_params_}")
    print("Test set performance:")
    evaluate_model(best_model, X_test, y_test)
```

```python
print("\n=== Model Comparison ===")
for name, model in best_models.items():
    print(f"\n{name}:")
    evaluate_model(model, X_test, y_test)
```
